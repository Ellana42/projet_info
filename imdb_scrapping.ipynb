```python
headers = {
    "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
    "accept-language": "en-GB,en;q=0.9,ja-JP;q=0.8,ja;q=0.7,fr-FR;q=0.6,fr;q=0.5,ru-RU;q=0.4,ru;q=0.3,en-US;q=0.2",
    "sec-fetch-dest": "document",
    "sec-fetch-mode": "navigate",
    "sec-fetch-site": "same-origin",
    "sec-fetch-user": "?1",
    "upgrade-insecure-requests": "1"
  }
```

```python
import urllib.parse
```

```python
def get_search(movie_name):
    query =  urllib.parse.quote(movie_name)
    url = 'https://www.imdb.com/find?q=' + query + '&ref_=nv_sr_sm'
    return url
```

```python
def get_movie(movie_name):
    url = get_search(movie_name)
    page = requests.get(url, headers=headers)
    return page.content

def get_movie_page(movie_name):
    search_page = get_movie(movie_name)
    search_soup = bs(search_page, 'html.parser')
    query = search_soup.find(id='pagecontent').find('tr').find('a')['href']
    url = 'https://www.imdb.com' + query + '?ref_=fn_al_tt_1'
    page = bs(requests.get(url).content, 'html.parser')
    return page
```

```python
lalapage = get_movie_page('La la land')
lalapage
```
